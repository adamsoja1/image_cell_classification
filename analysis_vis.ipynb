```python
import matplotlib.pyplot as plt

# Data
categories = ['Neoplastic', 'Connective tissue', 'Inflammatory', 'Epithelial', 'Dead']
counts = [72064, 47593, 30724, 24698, 2795]

# Plotting the bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, counts, color='skyblue')

# Adding labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 500, f'{yval}', ha='center', va='bottom')

# Adding titles and labels
plt.title('Counts by Category')
plt.xlabel('Category')
plt.ylabel('Count')

# Show the plot
plt.show()
```
