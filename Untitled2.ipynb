```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Your confusion matrix data
data = np.array([
    [735,  80, 179,   6],
    [109, 840,  48,   3],
    [134,  42, 816,   8],
    [ 12,  13,  11,  64]
])

# Labels for the axes
labels = ['Normal', 'Inflammatory', 'Tumor', 'Other']

# Create the heatmap
plt.figure(figsize=(8, 6))
ax = sns.heatmap(data, annot=True, fmt="d", cmap="YlGnBu", cbar=True, xticklabels=labels, yticklabels=labels)

# Add title and labels
ax.set_title('Confusion Matrix - ResNet-50', fontsize=16)
ax.set_xlabel('Predicted Labels', fontsize=12)
ax.set_ylabel('True Labels', fontsize=12)

# Display the heatmap
plt.show()
```
