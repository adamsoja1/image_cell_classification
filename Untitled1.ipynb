```python
import os
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from albumentations.pytorch import ToTensorV2
from torch.utils.data import DataLoader, Dataset
from albumentations import (
    Compose,
    OneOf,
    RandomBrightness,
    RandomContrast,
    MotionBlur,
    MedianBlur,
    GaussianBlur,
    VerticalFlip,
    HorizontalFlip,
RandomBrightnessContrast
)

# Define the transform with augmentation
transform = Compose(
    [
        OneOf([RandomBrightness(limit=0.1, p=1), RandomContrast(limit=0.1, p=0.8)]),
        OneOf([MotionBlur(blur_limit=3), MedianBlur(blur_limit=3), GaussianBlur(blur_limit=3)], p=0.7),
        VerticalFlip(p=0.5),
        HorizontalFlip(p=0.5),
        ToTensorV2()
    ]
)

# ImageDataset class as provided
class ImageDataset(Dataset):
    def __init__(self, data_path, transform=None):
        self.transform = transform
        self.dataset = self.load_dataset(data_path)

    def load_dataset(self, path):
        files = os.listdir(path)
        dataset_final = pd.DataFrame(columns=['filename', 'class'])
        for filename in files:
            if filename.endswith('.txt'):
                files = self.get_images_list(f'{path}/{filename}')
                dataset = pd.DataFrame({
                    'filename': files,
                    'class': filename.split('_')[1][:-3]
                })
                dataset_final = pd.concat([dataset_final, dataset], ignore_index=True)
        return dataset_final

    def get_images_list(self, file_path):
        # Assuming the text file contains image file names
        with open(file_path, 'r') as file:
            image_files = file.read().splitlines()
        return image_files

    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        image = cv2.imread(f'{self.dataset["filename"].loc[idx]}')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        #image = cv2.resize(image, (32, 32), interpolation=cv2.INTER_CUBIC)
        image = resize_with_padding(image, (32, 32))

        if self.transform:
            image = self.transform(image=image)['image']
        else:
            image = ToTensorV2()(image=image)['image']
        
        target = self.dataset["class"].loc[idx]
        return image, target

# Load the dataset
data_path = 'train_data'  # Replace with the path to your dataset
dataset_with_aug = ImageDataset(data_path, transform=transform)
dataset_without_aug = ImageDataset(data_path, transform=None)

# DataLoader to get a batch of 16 images
batch_size = 16
loader_with_aug = DataLoader(dataset_with_aug, batch_size=batch_size, shuffle=False)
loader_without_aug = DataLoader(dataset_without_aug, batch_size=batch_size, shuffle=False)

# Get a batch

images_with_aug, _ = next(iter(loader_with_aug))
images_without_aug, _ = next(iter(loader_without_aug))

# Plot the images in subplots
fig, axes = plt.subplots(8, 4, figsize=(6, 12))  # 8 rows, 4 columns

for i in range(batch_size):
    # Without augmentation
    ax = axes[i // 2, (i % 2) * 2]
    ax.imshow(np.transpose(images_without_aug[i].numpy(), (1, 2, 0)))
    ax.set_title('Original')
    ax.axis('off')
    
    # With augmentation
    ax = axes[i // 2, (i % 2) * 2 + 1]
    ax.imshow(np.transpose(images_with_aug[i].numpy(), (1, 2, 0)))
    ax.set_title('Augmented')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from utils_cells import get_images_list, transform_image, transform_target, resize_with_padding
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
import numpy as np
import torchvision.transforms.functional as F
import torch
from torchvision import transforms
from torchvision.transforms import functional as F
import cv2
from sklearn.model_selection import train_test_split
import optuna
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.models import resnet18
from torchmetrics import Precision, Recall
import numpy as np
import datetime
import random
import time
import torchvision.models as models

import random

class ImageDataset(Dataset):
    def __init__(self, data_path, transform=None, target_transform=None, reduce=False):
        self.transform = transform
        self.target_transform = target_transform
        self.dataset = shuffle(self.load_dataset(data_path))

    def load_dataset(self, path):
        path = []
        classes = []
        for image_class in os.listdir('cells_final'):
            for img in os.listdir(f'cells_final/{image_class}'):
                path.append(f'cells_final/{image_class}/{img}')
                classes.append(image_class)

        dataset_final = pd.DataFrame()
        dataset_final['path'] = path
        dataset_final['class'] = classes
        return dataset_final                
                          
    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        image = cv2.imread(f'{self.dataset["path"].loc[idx]}')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (32, 32), interpolation=cv2.INTER_CUBIC)
        
        #image = resize_with_padding(image, (32, 32))
        image = image.astype(np.float32)
        image = image/255.0
        image = self.transform(image = image)['image'] if self.transform is not None else image

        target = self.dataset['class'].loc[idx]

        if target.strip() == 'normal':
            target_ = [1, 0, 0, 0]
        elif target.strip() == 'inflammatory':
            target_ = [0, 1, 0, 0]
        elif target.strip() == 'tumor':
            target_ = [0, 0, 1, 0]
        elif target.strip() == 'other':
            target_ = [0, 0, 0, 1]
        else:
            print(target)
        
        image = F.to_tensor(image)
        

        return image.float(), torch.Tensor(np.array(target_, dtype=np.float32))
```

```python
import matplotlib.pyplot as plt

# Initialize the dataset
dataset = ImageDataset(data_path='cells_final', transform=None)

# Initialize a dictionary to store images by class
class_images = {'normal': [], 'inflammatory': [], 'tumor': [], 'other': []}

# Group images by their class
for i in range(len(dataset)):
    image, target = dataset[i]
    class_label = dataset.dataset['class'].loc[i].strip()
    if len(class_images[class_label]) < 4:
        class_images[class_label].append(image)
    # Break early if we have enough images
    if all(len(images) == 4 for images in class_images.values()):
        break

# Plotting with smaller images
fig, axes = plt.subplots(4, 4, figsize=(6, 6))  # Smaller figsize for smaller images

# Iterate through each class and its corresponding images
for idx, (class_label, images) in enumerate(class_images.items()):
    for j in range(4):
        ax = axes[idx, j]
        ax.imshow(np.transpose(images[j].numpy(), (1, 2, 0)))
        ax.axis('off')
        ax.text(2, 30, class_label.capitalize(), color='white', fontsize=8, 
                bbox=dict(facecolor='black', alpha=0.5, pad=2))

plt.subplots_adjust(wspace=0.1, hspace=0.1)  # Reduce space between plots
plt.show()
```
